### Proper example !

```python
# @njit
def make_signal(long_enter=False, long_exit=False, short_enter=False, short_exit=False):
     return long_enter, long_exit, short_enter, short_exit

# @njit
def signal_func_nb(c, fast, slow):
    if vbt.nb.iter_crossed_above_nb(fast, slow, c.i, c.col):
         return make_signal(long_enter=True)
    if vbt.nb.iter_crossed_below_nb(fast, slow, c.i, c.col):
         return make_signal(short_enter=True)
    return make_signal()


fast = df.run('sma', 10, short_name='fast').real
slow = df.run('sma', 60, short_name='slow').real

pf = vbt.Portfolio.from_signals(
    close,
    signal_func_nb=signal_func_nb,
    signal_args=(
        vbt.Rep('fast'),
        vbt.Rep('slow'),
    ),
    broadcast_named_args=dict(
        fast=fast,
        slow=slow,
    ),
    size=1.0,
    size_type="valuepercent",
    init_cash=1000000.0,
    size_granularity=1,
    jitted=False,
)
pf
```

### Pipeline with all variations!

```python
symbols = ["MES", "6B"]
df = pth.SierraChartData.ffill(
    pth.SierraChartData.pull(
        symbols,
        timeframe="1min",
        start="2024-11-01 13:00",
        end="2024-12-01 20:00",
    )
)
df
```

```python
close = df.get('close')
close['MES'].vbt.plot()
```

```python
close.shape
```

```python
@njit
def make_signal(long_enter=False, long_exit=False, short_enter=False, short_exit=False):
     return long_enter, long_exit, short_enter, short_exit


@njit
def signal_func_nb(c, fast, slow):
    if vbt.nb.iter_crossed_above_nb(fast, slow, c.i, c.col):
         return make_signal(long_enter=True)
    if vbt.nb.iter_crossed_below_nb(fast, slow, c.i, c.col):
         return make_signal(short_enter=True)
    return make_signal()


def pipeline(close, fast=20, slow=60, size=1.):
    fast = df.run('sma', fast, short_name='fast').real
    slow = df.run('sma', slow, short_name='slow').real

    pf = vbt.Portfolio.from_signals(
        close,
        signal_func_nb=signal_func_nb,
        signal_args=(
            vbt.Rep('fast'),
            vbt.Rep('slow'),
        ),
        broadcast_named_args=dict(
            fast=fast,
            slow=slow,
        ),
        size=size,
        size_type="valuepercent",
        init_cash=1000000.0,
        # size_granularity=1,
    )
    return pf
```

```python
pf = pipeline(
    close,
    fast=np.arange(10, 200, 5),
    slow=np.arange(60, 1200, 30),
    size=vbt.Param([.25, .5, 1.])
)
pf
```
